```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("csv_xlsx/Tata_Retail.csv")
df
```

```python
df= df.drop_duplicates()
df=df.dropna(subset=['InvoiceNo', 'StockCode', 'Description', 'Quantity',
                     'InvoiceDate', 'UnitPrice', 'Country'])
df['InvoiceDate']= pd.to_datetime(df['InvoiceDate'], dayfirst=True,
                                   errors= 'coerce')
df = df.drop(df.columns[0], axis=1)
df=df.dropna(subset=['InvoiceDate'])
df['Description']= df['Description'].str.strip()
df['Country']=df['Country'].str.strip()
sales=df[df['Quantity']>0].copy()
returns=df[df['Quantity']<0].copy()
sales= sales.reset_index(drop=True)
returns= returns.reset_index(drop=True)
sales.to_csv("csv_xlsx/sales_data.csv", index= False)
sales.to_excel("csv_xlsx/sales_data.xlsx", index= False)
returns.to_csv("csv_xlsx/returns_data.csv", index= False)
returns.to_excel("csv_xlsx/returns_data.xlsx", index= False)
df
```

```python
sales= pd.read_csv('csv_xlsx/sales_data.csv', dtype={'InvoiceNo': str}, low_memory=False)
returns= pd.read_csv('csv_xlsx/returns_data.csv', dtype={'InvoiceNo': str}, low_memory=False)

sales['InvoiceDate']= pd.to_datetime(sales['InvoiceDate'])
returns['InvoiceDate']= pd.to_datetime(returns['InvoiceDate'])

#Creating revenue and returnvalue columns
sales['Revenue']= sales['Quantity']*sales['UnitPrice']
returns['ReturnValue']= returns['Quantity']*returns['UnitPrice'] #negative

#Daily sales and returns
daily_sales= sales.groupby(sales['InvoiceDate'].dt.date)['Revenue'].sum().reset_index()
daily_returns= returns.groupby(returns['InvoiceDate'].dt.date)['ReturnValue'].sum().reset_index()

#Monthly sales and returns
monthly_sales= sales.groupby(sales['InvoiceDate'].dt.to_period('M'))['Revenue'].sum().reset_index()
monthly_returns= returns.groupby(returns['InvoiceDate'].dt.to_period('M'))['ReturnValue'].sum().reset_index()

#Export for Tableau
daily_sales.to_csv('csv_xlsx/tableau_daily_sales.csv', index=False)
daily_sales.to_excel('csv_xlsx/tableau_daily_sales.xlsx', index=False)
daily_returns.to_csv('csv_xlsx/tableau_daily_returns.csv', index=False)
daily_returns.to_excel('csv_xlsx/tableau_daily_returns.xlsx', index=False)
monthly_sales.to_csv('csv_xlsx/tableau_monthly_sales.csv', index=False)
monthly_sales.to_excel('csv_xlsx/tableau_monthly_sales.xlsx', index=False)
monthly_returns.to_csv('csv_xlsx/tableau_monthly_returns.csv', index=False)
monthly_returns.to_excel('csv_xlsx/tableau_monthly_returns.xlsx', index=False)

#TOP PRODUCTS
top_sales_products= sales.groupby('Description')['Revenue'].sum().sort_values(ascending=False).reset_index()
top_return_products= returns.groupby('Description')['ReturnValue'].sum().sort_values().reset_index()

#Export for Tableau
top_sales_products.to_csv('csv_xlsx/tableau_top_sales_products.csv', index=False)
top_sales_products.to_excel('csv_xlsx/tableau_top_sales_products.xlsx', index=False)
top_return_products.to_csv('csv_xlsx/tableau_top_returns_products.csv', index=False)
top_return_products.to_excel('csv_xlsx/tableau_top_returns_products.xlsx', index=False)

#SALES & RETURNS BY COUNTRY
sales_by_country= sales.groupby('Country')['Revenue'].sum().sort_values(ascending=False).reset_index()
returns_by_country= returns.groupby('Country')['ReturnValue'].sum().sort_values().reset_index()
#Export for tableau
sales_by_country.to_csv('csv_xlsx/tableau_sales_by_country.csv', index=False)
sales_by_country.to_excel('csv_xlsx/tableau_sales_by_country.xlsx', index=False)
returns_by_country.to_csv('csv_xlsx/tableau_returns_by_country.csv', index=False)
returns_by_country.to_excel('csv_xlsx/tableau_returns_by_country.xlsx', index=False)
#Order Size and Value Distribution
order_value=sales.groupby('InvoiceNo').agg(
    OrderValue=('Revenue', 'sum'),
    OrderSize=('Quantity', 'sum')
).reset_index()
#Export for tableau
order_value.to_csv('csv_xlsx/tableau_order_value_size.csv', index=False)
order_value.to_excel('csv_xlsx/tableau_order_value_size.xlsx', index=False)
```

```python
#1. Daily Sales and returns Line Plot
plt.figure(figsize=(12,6))
plt.plot(daily_sales['InvoiceDate'], daily_sales['Revenue'], label='Daily Sales')
plt.plot(daily_returns['InvoiceDate'], daily_returns['ReturnValue'], label='Daily Returns')
plt.title('Daily Sales vs Returns')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Outputs/Daily_Sales_vs_Returns.png", dpi=300, bbox_inches='tight')
plt.show()

#2. Monthly sales and returns bar chart
plt.figure(figsize=(10,5))
plt.bar(monthly_sales['InvoiceDate'].astype(str), monthly_sales['Revenue'],
    label='Monthly Sales')
plt.bar(monthly_returns['InvoiceDate'].astype(str), monthly_returns['ReturnValue'],
        label= 'Monthly Returns')
plt.title('Monthly Sales vs Returns')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Outputs/Mothly_Sales_vs_Returns.png", dpi=300, bbox_inches='tight')
plt.show()

#3. Top 10 products sold
plt.figure(figsize=(10,6))
top10_sales=top_sales_products.head(10)
plt.barh(top10_sales['Description'][::-1], top10_sales['Revenue'][::-1])
plt.title('Top 10 Products by Sales Revenue')
plt.xlabel('Revenue')
plt.tight_layout()
plt.savefig("Outputs/Top_10_Products_By_Revenue.png", dpi=300, bbox_inches='tight')
plt.show()

#4. Top 10 Products Returned
plt.figure(figsize=(10,6))
top10_returns=top_return_products.head(10)
plt.barh(top10_returns['Description'][::-1], top10_returns['ReturnValue'][::-1])
plt.title('Top 10 Products by Returns (absolute value)')
plt.xlabel('Return Value')
plt.tight_layout()
plt.savefig("Outputs/Top_10_Products_by_Return.png", dpi=300, bbox_inches='tight')
plt.show()

#5. Sales and returns by country (top10)
plt.figure(figsize=(10,6))
top_countries_sales= sales_by_country.head(10)
plt.barh(top_countries_sales['Country'][::-1], top_countries_sales['Revenue'][::-1])
plt.title('Top 10 Countries by Sales Revenue')
plt.xlabel('Revenue')
plt.tight_layout()
plt.savefig("Outputs/Sales_by_country.png", dpi=300, bbox_inches='tight')
plt.show()
plt.figure(figsize=(10,6))
top_countries_returns= returns_by_country.head(10)
plt.barh(top_countries_returns['Country'][::-1], top_countries_returns['ReturnValue'][::-1])
plt.title('Top 10 Countries by Returns(absolute value)')
plt.xlabel('Return Value')
plt.tight_layout()
plt.savefig("Outputs/Top_10_Countries_by_returns.png", dpi=300, bbox_inches='tight')
plt.show()

#6. Order Value Distribution (histogram)
plt.figure(figsize=(10,6))
plt.hist(order_value['OrderValue'], bins=30)
plt.title('Distribution of Order Values')
plt.xlabel('Order Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig("Outputs/Dist_Order_Values.png", dpi=300, bbox_inches='tight')
plt.show()

#7.Order Size Distribution
plt.figure(figsize=(10,6))
plt.hist(order_value['OrderSize'], bins=30)
plt.title('Distribution of Order Sizes')
plt.xlabel('Order Size')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig("Outputs/Dist_Order_Sizes.png", dpi=300, bbox_inches='tight')
plt.show()
```
